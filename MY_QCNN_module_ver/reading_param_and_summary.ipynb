```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
```

```python
## reading param
file_path = \
"/content/drive/MyDrive/박경덕교수님연구실/MY_QCNN_module_ver/pca_Amplitude_ConvCirc1_mse_result"
param_list = []
for i in range(200):
  param_list.append(np.load(file_path+"/"+f"iter{i}.npy"))
param_list = np.array(param_list)
```

```python
## reading loss
import json
with open(file_path+"/summary.json", "r") as f:
  summary_dict = json.load(f)
f.close()
total_time = summary_dict["total_time"]
loss_list = summary_dict["loss"]
print(f"Total training time: {total_time}")
print(f"Test accuracy: {summary_dict['accuracy']}")
```

```python
import matplotlib.pyplot as plt
plt.plot(loss_list)
plt.title(file_path.split("/")[-1])
plt.xlabel("ITER")
plt.ylabel("LOSS")
plt.ylim((0.4, 1.0))

plt.show()
```
