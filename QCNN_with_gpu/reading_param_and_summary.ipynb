```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
```

```python
### enter the filepath where you want to load files
file_path = \
"/content/drive/MyDrive/박경덕교수님연구실/QCNN_with_gpu/pca_Angle-compact_U_5_cross_entropy_result"
```

```python
### loading parameter history
param_list = []
for i in range(200):
  param_list.append(np.load(file_path+"/"+f"param_of_iter{i}.npy", allow_pickle=True))
param_list = np.array(param_list)
```

```python
## reading summary
import json
with open(file_path+"/summary.json", "r") as f:
  summary_dict = json.load(f)
f.close()
total_time = summary_dict["total_time"]
loss_list = summary_dict["loss"]
print(f"Total training time: {total_time}")
print(f"Test accuracy: {summary_dict['accuracy']}")
```

```python
### drawing loss plot
import matplotlib.pyplot as plt
plt.plot(loss_list)
plt.title(file_path.split("/")[-1])
plt.xlabel("ITER")
plt.ylabel("LOSS")
plt.ylim((0.35, 0.75))

plt.show()
```

```python
### history of parameters
plt.figure(figsize = (20,10))
plt.plot(param_list)
plt.title("history of parameters")
plt.xlabel("ITER")
plt.ylabel("PARAMETERS")
plt.show()
```
